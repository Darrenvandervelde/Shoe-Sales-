# Shoe Sales Data Analysis (Yearly)
#
### This notebook analyzes a small shoe sales dataset stored in an array.
### It calculates total revenue, monthly revenue, revenue by shoe type, and visualizes the data.

```python
# --- Import Libraries ---
import matplotlib.pyplot as plt
```

```python

# --- Sales Data ---
# Format: [Month, Shoe Name, Units Sold, Price Per Unit]
sales_data = [
    ["January", "Running Shoes", 120, 60],
    ["February", "Sneakers", 150, 55],
    ["March", "Boots", 90, 80],
    ["April", "Running Shoes", 130, 60],
    ["May", "Sneakers", 170, 55],
    ["June", "Sandals", 200, 40],
    ["July", "Sandals", 220, 40],
    ["August", "Sneakers", 160, 55],
    ["September", "Boots", 110, 80],
    ["October", "Running Shoes", 140, 60],
    ["November", "Sneakers", 190, 55],
    ["December", "Boots", 150, 80],
]
```

```python
# --- Print Sales Data ---
print("Sales Data:\n")
for row in sales_data:
    print(row)
```

```python
# --- Calculate Total Yearly Revenue ---
total_revenue = sum(units * price for month, shoe, units, price in sales_data)
print(f"\nTotal Yearly Revenue: ${total_revenue}")
```

```python
# --- Revenue by Shoe Type ---
shoe_revenue = {}
for month, shoe, units, price in sales_data:
    if shoe not in shoe_revenue:
        shoe_revenue[shoe] = 0
    shoe_revenue[shoe] += units * price
```

```python
print("\nRevenue by Shoe Type:")
for shoe, revenue in shoe_revenue.items():
    print(f"{shoe}: ${revenue}")
```

```python
# --- Monthly Revenue Summary ---
print("\nMonthly Revenue Summary:")
for month, shoe, units, price in sales_data:
    print(f"{month}: ${units * price}")
```

```python
# --- Key Insights ---
print("\nKey Insights:")
print("- Sneakers generated strong sales across multiple months")
print("- Sandals performed best during summer months")
print("- Boots peaked toward the end of the year")
print("- Overall yearly revenue shows consistent demand")
```

```python
# --- Visualization: Monthly Revenue Bar Chart ---
months = [row[0] for row in sales_data]
monthly_revenue = [row[2] * row[3] for row in sales_data]
```

```python

plt.figure(figsize=(12, 6))
plt.bar(months, monthly_revenue, color='skyblue')
plt.title('Monthly Revenue', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# --- Visualization: Revenue by Shoe Type Pie Chart ---
shoe_names = list(shoe_revenue.keys())
revenues = list(shoe_revenue.values())
```

```python
plt.figure(figsize=(8, 8))
plt.pie(revenues, labels=shoe_names, autopct='%1.1f%%', startangle=140,
        colors=['#FF9999','#66B3FF','#99FF99','#FFCC99'])
plt.title('Revenue by Shoe Type', fontsize=16)
plt.show()
```
